```python
# Imaginons une liste d'utilisateurs avec des âges manquants (None)
users = pd.DataFrame({
    'nom': ['Safidy', 'Antonio', 'Vanissa', 'Inconnu'],
    'age': [25, 30, None, 40],
    'achat': ['Oui', 'Non', 'Oui', 'Non']
})

# 1. Voir les premières lignes
print(users.head())

# 2. Remplir l'âge manquant par la moyenne d'âge
users['age'] = users['age'].fillna(users['age'].mean())

# 3. Transformer 'Oui'/'Non' en 1/0 (Encoding)
users['achat'] = users['achat'].map({'Oui': 1, 'Non': 0})

print("\nDonnées nettoyées :")
print(users)
```
