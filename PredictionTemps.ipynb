```python
import numpy as np
from sklearn.linear_model import LinearRegression

# 1. Données d'entraînement (X = nb de mots, y = heures passées)
# On utilise .reshape(-1, 1) car scikit-learn attend un tableau 2D pour X
X = np.array([10, 50, 100, 150, 200, 300]).reshape(-1, 1)
y = np.array([2, 5, 12, 18, 22, 35])

# 2. Création et entraînement du modèle
model = LinearRegression()
model.fit(X, y)

# 3. Prédiction pour un nouveau ticket de 120 mots
nouveau_ticket = np.array([[120]])
estimation = model.predict(nouveau_ticket)

print(f"Estimation pour 120 mots : {estimation[0]:.2f} heures")
```
