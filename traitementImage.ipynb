```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# 1. Ouvrir l'image locale (celle que tu as glissée dans Colab)
path = "/content/Capture d'écran bouton retour .png"
img = Image.open(path)

# 2. Transformer l'image en tableau de chiffres (Numpy)
img_array = np.array(img)

# 3. Afficher les dimensions
# Tu verras peut-être un '4' à la fin si l'image a de la transparence (RGBA)
print(f"Dimensions de l'image : {img_array.shape}")

# 4. Afficher l'image
plt.imshow(img_array)
plt.axis('off') # Pour enlever les graduations
plt.show()
```
